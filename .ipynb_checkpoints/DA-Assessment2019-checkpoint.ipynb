# Statistical analysis 
is a process of understanding how variables in a dataset relate to each other and how those relationships depend on other variables. 

```python
sns.scatterplot(x='total_bill',y='tip',data = tips)
```

```python
g = sns.FacetGrid(tips, col="time",  row="smoker")
g = g.map(plt.hist, "total_bill") #reference https://seaborn.pydata.org/generated/seaborn.FacetGrid.html
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex")
g = g.map(plt.hist, "total_bill")
```
